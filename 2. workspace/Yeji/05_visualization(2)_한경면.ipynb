```python
df = pd.read_csv('./datas/df3.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

### 한경면

```python
hankyung = df[df['dong'].isin(['한경면'])]
hankyung
```

```python
hankyung_type = hankyung.groupby(['YM', 'Type'], as_index=False).mean()
hankyung_type[hankyung_type['YM'] == 202008].sort_values(by='dis/tot', ascending=False).head(10)
```

```python
col = ['NumofSpent', 'NumofDisSpent']
num_hankyung_5 = (hankyung_type[hankyung_type['YM'] == 202005][col]) / (hankyung_type[hankyung_type['YM'] == 202005][col].max())
num_hankyung_6 = (hankyung_type[hankyung_type['YM'] == 202006][col]) / (hankyung_type[hankyung_type['YM'] == 202006][col].max())
num_hankyung_7 = (hankyung_type[hankyung_type['YM'] == 202007][col]) / (hankyung_type[hankyung_type['YM'] == 202007][col].max())
num_hankyung_8 = (hankyung_type[hankyung_type['YM'] == 202008][col]) / (hankyung_type[hankyung_type['YM'] == 202008][col].max())
```

```python
norm_hankyung_5 = hankyung_type[hankyung_type['YM'] == 202005].copy()
norm_hankyung_6 = hankyung_type[hankyung_type['YM'] == 202006].copy()
norm_hankyung_7 = hankyung_type[hankyung_type['YM'] == 202007].copy()
norm_hankyung_8 = hankyung_type[hankyung_type['YM'] == 202008].copy()

norm_hankyung_5[col] = num_hankyung_5[col]
norm_hankyung_6[col] = num_hankyung_6[col]
norm_hankyung_7[col] = num_hankyung_7[col]
norm_hankyung_8[col] = num_hankyung_8[col]
```

```python
norm_hankyung_5.nlargest(10, 'dis/tot')
```

```python
fig = plt.figure(figsize=(16, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax1 = sns.set_color_codes('muted')
ax1 = sns.barplot(x='Type', y='NumofDisSpent', data=norm_hankyung_5.nlargest(10, 'dis/tot'), color='b')
ax1.set_title('5월 업종별 평균 재난지원금 이용건수')
plt.setp(ax1.get_xticklabels(), rotation=45)
ax1.set(ylim=(0.0, 1.0))
ax2 = fig.add_subplot(1, 2, 2)
ax2 = sns.set_color_codes('pastel')
ax2 = sns.barplot(x='Type', y='NumofSpent', data=norm_hankyung_5.nlargest(10, 'dis/tot'), color='b')
ax2.set_title('5월 업종별 평균 이용건수')
plt.setp(ax2.get_xticklabels(), rotation=45)
ax2.set(ylim=(0.0, 1.0))
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster, MiniMap
```

```python
map = folium.Map((33.32085488754027, 126.24562415185463), zoom_start=13)
mc = MarkerCluster()

mc.add_child(folium.Marker(location=[33.32085488754027, 126.24562415185463], popup='중소2', icon=folium.Icon(color='red')))
minimap = MiniMap()
map.add_child(minimap)
```

```python
# 농축협직영매장
nonghyop = hankyung[hankyung['Type'] == '농축협직영매장']
nonghyop = nonghyop[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
nonghyop.drop_duplicates(inplace=True)

# 골프경기장
golf = hankyung[hankyung['Type'] == '골프경기장']
golf = golf[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
golf.drop_duplicates(inplace=True)

# 슈퍼마켓
mart = hankyung[hankyung['Type'] == '슈퍼마켓']
mart = mart[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
mart.drop_duplicates(inplace=True)

# 비료농약사료종자
biryo = hankyung[hankyung['Type'] == '비료농약사료종자']
biryo = biryo[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
biryo.drop_duplicates(inplace=True)

# 주유소
gas = hankyung[hankyung['Type'] == '주유소']
gas = gas[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
gas.drop_duplicates(inplace=True)
```

```python
map = folium.Map((33.32085488754027, 126.24562415185463), zoom_start= 13)

# 농축협직영매장
for i in range(len(nonghyop['Type'])):
    folium.Circle(list(nonghyop.iloc[i][['lat','lon']]), radius=100, color='crimson', fill_color='crimson').add_to(map)
    
# 골프경기장 
for i in range(len(golf['Type'])):
    folium.Circle(list(golf.iloc[i][['lat','lon']]), radius=100, color='orange', fill_color='orange').add_to(map)
    
# 슈퍼마켓 
for i in range(len(mart['Type'])):
    folium.Circle(list(mart.iloc[i][['lat','lon']]), radius=100, color='green', fill_color='green').add_to(map)

# 비료농약사료종자 
for i in range(len(biryo['Type'])):
    folium.Circle(list(biryo.iloc[i][['lat','lon']]), radius=100, color='gray', fill_color='gray').add_to(map)
    
# 주유소 
for i in range(len(gas['Type'])):
    folium.Circle(list(gas.iloc[i][['lat','lon']]), radius=100, color='black', fill_color='gray').add_to(map)
    
map
```
