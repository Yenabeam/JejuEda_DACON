```python
df = pd.read_csv('./datas/df3.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

### 애월읍

```python
ewol = df[df['dong'].isin(['애월읍'])]
ewol
```

```python
ewol_type = ewol.groupby(['YM', 'Type']).mean()
ewol_type.reset_index(inplace=True)
ewol_type[ewol_type['YM'] == 202005].sort_values(by='dis/tot', ascending=False).head(10)
```

```python
col = ['NumofSpent', 'NumofDisSpent']
num_ewol_5 = (ewol_type[ewol_type['YM'] == 202005][col]) / (ewol_type[ewol_type['YM'] == 202005][col].max())
num_ewol_6 = (ewol_type[ewol_type['YM'] == 202006][col]) / (ewol_type[ewol_type['YM'] == 202006][col].max())
num_ewol_7 = (ewol_type[ewol_type['YM'] == 202007][col]) / (ewol_type[ewol_type['YM'] == 202007][col].max())
num_ewol_8 = (ewol_type[ewol_type['YM'] == 202008][col]) / (ewol_type[ewol_type['YM'] == 202008][col].max())
```

```python
norm_ewol_5 = ewol_type[ewol_type['YM'] == 202005].copy()
norm_ewol_6 = ewol_type[ewol_type['YM'] == 202006].copy()
norm_ewol_7 = ewol_type[ewol_type['YM'] == 202007].copy()
norm_ewol_8 = ewol_type[ewol_type['YM'] == 202008].copy()

norm_ewol_5[col] = num_ewol_5[col]
norm_ewol_6[col] = num_ewol_6[col]
norm_ewol_7[col] = num_ewol_7[col]
norm_ewol_8[col] = num_ewol_8[col]
```

```python
norm_ewol_5.nlargest(10, 'dis/tot')
```

```python
fig = plt.figure(figsize=(16, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax1 = sns.set_color_codes('muted')
ax1 = sns.barplot(x='Type', y='NumofDisSpent', data=norm_ewol_5.nlargest(10, 'dis/tot'), color='b')
ax1.set_title('5월 업종별 평균 재난지원금 이용건수')
plt.setp(ax1.get_xticklabels(), rotation=45)
ax1.set(ylim=(0.0, 0.030))
ax2 = fig.add_subplot(1, 2, 2)
ax2 = sns.set_color_codes('pastel')
ax2 = sns.barplot(x='Type', y='NumofSpent', data=norm_ewol_5.nlargest(10, 'dis/tot'), color='b')
ax2.set_title('5월 업종별 평균 이용건수')
plt.setp(ax2.get_xticklabels(), rotation=45)
ax2.set(ylim=(0.0, 0.030))
plt.show()
```

```python
# 농축협직영매장
nonghyop = ewol[ewol['Type'] == '농축협직영매장']
nonghyop = nonghyop[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
nonghyop.drop_duplicates(inplace=True)

# 기타잡화
others = ewol[ewol['Type'] == '기타잡화']
others = others[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
others.drop_duplicates(inplace=True)

# 기타농업관련
othernong = ewol[ewol['Type'] == '기타농업관련']
othernong = othernong[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
othernong.drop_duplicates(inplace=True)

# 조세서비스
jose = ewol[ewol['Type'] == '조세서비스']
jose = jose[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
jose.drop_duplicates(inplace=True)

# # 주유소
# gas = ewol[ewol['Type'] == '주유소']
# gas = gas[['SIGUNGU','FranClass','Type','lon','lat','lon_lat']]
# gas.drop_duplicates(inplace=True)
```

```python
map = folium.Map((33.447843033444606, 126.30114679770172), zoom_start= 13)

# 농축협직영매장
for i in range(len(nonghyop['Type'])):
    folium.Circle(list(nonghyop.iloc[i][['lat','lon']]), radius=100, color='crimson', fill_color='crimson').add_to(map)
    
# 기타잡화 
for i in range(len(others['Type'])):
    folium.Circle(list(others.iloc[i][['lat','lon']]), radius=100, color='orange', fill_color='orange').add_to(map)
    
# 기타농업관련 
for i in range(len(othernong['Type'])):
    folium.Circle(list(othernong.iloc[i][['lat','lon']]), radius=100, color='green', fill_color='green').add_to(map)

# 조세서비스
for i in range(len(jose['Type'])):
    folium.Circle(list(jose.iloc[i][['lat','lon']]), radius=100, color='gray', fill_color='gray').add_to(map)
    
# # 주유소 
# for i in range(len(gas['Type'])):
#     folium.Circle(list(gas.iloc[i][['lat','lon']]), radius=100, color='black', fill_color='gray').add_to(map)
    
map
```
